```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, morphology

import pyclesperanto_prototype as cle

from pyclesperanto_prototype import execute

from pyclesperanto_prototype import plugin_function
from pyclesperanto_prototype import Image
```

```python
image = np.zeros((1, 10, 10))
image[0, 5, 5] = 1
```

```python
cle_image = cle.push(image)
```

```python
@plugin_function(categories=['filter', 'edge detection', 'in assistant'])
def gradient(source: Image, destination: Image = None):

    parameters = {
            "src":source,
            "dst":destination
        }
    
    execute(None, 'gradient_x.cl', 'gradient_x', destination.shape, parameters)
    return destination
```

```python
gradient(cle_image)
```
